# **Visualization**

```python
import utils
import os
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
datasets = ['cifar10', 'cifar100', 'mnist']
optimizers = ['SGD', 'ADAM', 'adahessian', 'Sophia']
colors = ['#E87B9F', '#A46750', '#479A5F', '#5BC5DB']
names = ['SGD', 'Adam', 'AdaHessian', 'Sophia']
```

```python
def get_generalization_error(logs):
   gen_array = []
   for log in logs:
      gen_array.append(float(log['train_acc']) - float(log['test_acc']))
   return gen_array
```

```python
def finding_best_hyper():
  out = dict()
  for d in datasets:
    for opt in optimizers:
      dir = f"log/{d}/{opt}"
      learning_rates = [float(name.split('_')[1]) for name in os.listdir(dir)]
      valid_accs = []
      for lr in learning_rates:
        path = dir + f"/lr_{lr}" + "_momentom_0.9_weightdecay_0.0002/logs.pkl"
        logs = utils.load_logs(path)
        valid_accs.append(logs[-1]['val_acc'])
      best_lr = learning_rates[np.argmax(valid_accs)]
      print(f'The best learning rate for {d}, {opt} is {best_lr}')
      out[d, opt] = best_lr
  return out
```

```python
best_lr = finding_best_hyper()
```

```python
def plot_gen_error(dataset, best_lr):
    d = dataset
    for num, opt in enumerate(optimizers):
        dir = f"log/{d}/{opt}"
        lr = best_lr[d, opt]
        path = dir + f"/lr_{lr}" + "_momentom_0.9_weightdecay_0.0002/logs.pkl"
        logs = utils.load_logs(path)
        g = get_generalization_error(logs)
        plt.ylabel('Gen. Error')
        plt.xlabel('#epochs')
        if num in [3,2]:
          plt.plot(g, label = names[num], color=colors[num],
                   linewidth=2, linestyle='--')
        else:
          plt.plot(g, label = names[num], color=colors[num],
                   linewidth=2)
        # plt.legend()
```

```python
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Arial'] + plt.rcParams['font.serif']
```

```python
plt.style.use('ggplot')
font = {
        'weight' : 'bold',
        'size'   : 16}
matplotlib.rc('font', **font)
plt.figure(figsize=(6,4))
plot_gen_error(datasets[0], best_lr)
plt.tight_layout()
plt.savefig('gen_error_cifar10', format='pdf')
```

```python
plt.figure(figsize=(6,4))
plot_gen_error(datasets[1], best_lr)
plt.tight_layout()
plt.savefig('gen_error_cifar100', format='pdf')
```

```python
plt.figure(figsize=(6,4))
plot_gen_error(datasets[2], best_lr)
plt.tight_layout()
plt.savefig('gen_error_mnist', format='pdf')
```

```python
def plot_test_acc(dataset, best_lr):
    d = dataset
    for num, opt in enumerate(optimizers):
        dir = f"log/{d}/{opt}"
        lr = best_lr[d, opt]
        path = dir + f"/lr_{lr}" + "_momentom_0.9_weightdecay_0.0002/logs.pkl"
        logs = utils.load_logs(path)
        g = [float(log['test_acc']) for log in logs]
        plt.ylabel('Test Accuracy')
        plt.xlabel('#epochs')
        if num in [3,2]:
          plt.plot(g, label = names[num], color=colors[num],
                   linewidth=2, linestyle='--')
        else:
          plt.plot(g, label = names[num], color=colors[num],
                   linewidth=2)
        plt.legend()
```

```python
plt.figure(figsize=(6,4))
plot_test_acc(datasets[0], best_lr)
plt.tight_layout()
plt.savefig('test_acc_cifar10', format='pdf')
```

```python
plt.figure(figsize=(6,4))
plot_test_acc(datasets[1], best_lr)
plt.tight_layout()
plt.savefig('test_acc_cifar100', format='pdf')
```

```python
plt.figure(figsize=(6,4))
plot_test_acc(datasets[2], best_lr)
plt.tight_layout()
plt.savefig('test_acc_mnist', format='pdf', bbox_inches = "tight")
```

```python
def plot_time(dataset, best_lr):
    d = dataset
    for num, opt in enumerate(optimizers):
        dir = f"log/{d}/{opt}"
        lr = best_lr[d, opt]
        path = dir + f"/lr_{lr}" + "_momentom_0.9_weightdecay_0.0002/logs.pkl"
        logs = utils.load_logs(path)
        g = [float(log['time']) for log in logs]
        plt.ylabel('Runtime')
        plt.xlabel('#epochs')
        if num in [3,2]:
          plt.plot(g, label = names[num], color=colors[num],
                   linewidth=2, linestyle='--')
        else:
          plt.plot(g, label = names[num], color=colors[num],
                   linewidth=2)
```

```python
plt.figure(figsize=(6,4))
plot_time(datasets[0], best_lr)
plt.tight_layout()
plt.savefig('time_cifar10', format='pdf', bbox_inches = "tight")
```

```python
plt.figure(figsize=(6,4))
plot_time(datasets[1], best_lr)
plt.tight_layout()
plt.savefig('time_cifar100', format='pdf', bbox_inches = "tight")
```

```python
plt.figure(figsize=(6,4))
plot_time(datasets[2], best_lr)
plt.tight_layout()
plt.savefig('time_mnist', format='pdf', bbox_inches = "tight")
```
